```python
#给字母表排序
#shuffle() 方法将序列的所有元素随机排序。括号里面可以是列表，字符串不可以,Python3返回值为None，python2无返回值
import string
import random
s = string.ascii_lowercase #所有小写字母(a-z)
#s=string.ascii_letters #所有大小写字母(a-z,A-Z)
#s=string.ascii_uppercase #所有大写字母(A-Z)
s1 = list(s)
# print(s1)

random.shuffle(s1)

print(s1)#这是打乱的26个字母


#冒泡排序 打乱的26个字母
def bubble_sort(string):
    for j in range(len(string)-1):
        for i in range(len(string)-j-1):
            if string[i] > string[i+1]:
                string[i],string[i+1] = string[i+1],string[i]
    return string

print(bubble_sort(s1))#冒泡排序排好的
```
